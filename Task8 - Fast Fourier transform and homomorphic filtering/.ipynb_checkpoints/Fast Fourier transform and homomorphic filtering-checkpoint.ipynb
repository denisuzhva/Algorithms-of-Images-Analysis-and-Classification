## Fast Fourier transform
### Theory

The main idea behind the algorithm of fast Fourier transform is the use of symmetrical properties of the discrete one.
It appears that if $H = 2^N$ for some integer $N$, we can perform recursive separation of even parts of the sums from the odd ones, which would look like as follows (for the 1-dimensional case, assuming $x = 2r$ for some $r$):
$$
\begin{equation}
    \begin{gathered}
        F(u) = \sum_{x=0}^{H-1} f(x) e^{-j2\pi ux/H} = \sum_{x=0}^{H/2-1} f(2r) e^{-j2\pi ur/ (H/2)} + e^{-j2\pi u/H} \sum_{x=0}^{H/2-1} f(2r+1) e^{-j2\pi ur/ (H/2)} = \\
        = F_\textrm{even}(u) + e^{-j2\pi u/H} F_\textrm{odd}(u).
    \end{gathered}
    \tag{1}
\end{equation}
$$
Since $H = 2^N$, we can repeat the process described above $N$ times.

The FFT algorithm thus allows to reduce complexity of the Fourier transform from $\mathcal{O}(N^2)$ to $\mathcal{O}(N \log_2 N)$.

### Code

#### Discrete Fourier transform

```python
import numpy as np


def get_dft(img, shift=False):
    h = img.shape[0]
    w = img.shape[1]
    img_dft = np.zeros_like(img, dtype=np.complex)
    img = img.astype(np.complex)
    
    x_arr = np.arange(h)
    y_arr = np.arange(w)
    
    x_mat = np.repeat(x_arr.reshape((h, 1)), w, 1)
    y_mat = np.repeat(y_arr.reshape((1, w)), h, 0)
    
    if shift:
        exp_val = (x_mat + y_mat) * np.pi * 1j
        img = np.multiply(img, np.exp(exp_val))
    
    for u in range(h):
        for v in range(w):
            exp_val = (u * x_mat / h + v * y_mat / w) * np.pi * 2j
            img_dft[u, v] = np.sum(np.multiply(img, np.exp(exp_val)))

    return img_dft
```

#### Fast Fourier transform 1D 

```python
import numpy as np


def get_fft1(arr, visual=False):
    n = len(arr)
    if n == 1:
        return arr
    arr_fft = [0] * n
    
    if visual:
        arr_ev, arr_od = get_fft1(arr[0::2], True), get_fft1(arr[1::2], True)
        
        for m in range(n//2):
            arr_fft[m] = arr_ev[m] + np.exp(-(np.pi * 2j * m) / n) * arr_od[m]
            arr_fft[m + n//2] = arr_ev[m] - np.exp(-(np.pi * 2j * m) / n) * arr_od[m]
            
        print('Part height: %i' % n)
        import matplotlib.pyplot as plt
        plt.figure(figsize=(10, 10), dpi=100)
        plt.matshow(np.log(np.abs(arr_fft)+1), cmap='gray')
        plt.show()
    else:
        arr_ev, arr_od = get_fft1(arr[0::2]), get_fft1(arr[1::2])
        
        for m in range(n//2):
            arr_fft[m] = arr_ev[m] + np.exp(-(np.pi * 2j * m) / n) * arr_od[m]
            arr_fft[m + n//2] = arr_ev[m] - np.exp(-(np.pi * 2j * m) / n) * arr_od[m]
    
        
    return arr_fft
```

#### Fast Fourier transform 2D 

```python
import numpy as np


def get_fft2(img, shift=False, visual=False):
    h = img.shape[0]
    w = img.shape[1]
    img = img.astype(np.complex)
    
    x_arr = np.arange(h)
    y_arr = np.arange(w)
    
    x_mat = np.repeat(x_arr.reshape((h, 1)), w, 1)
    y_mat = np.repeat(y_arr.reshape((1, w)), h, 0)
    
    if shift:
        exp_val = (x_mat + y_mat) * np.pi * 1j
        img = np.multiply(img, np.exp(exp_val))
    
    if visual:
        img_fft_0 = get_fft1(img, True)
    else:
        img_fft_0 = get_fft1(img)
    
    img_fft = np.transpose(get_fft1(np.transpose(img_fft_0)))
    return img_fft
```

#### Inverse fast Fourier transform 1D

```python
import numpy as np


def get_ifft1(arr_fft):
    arr_ifft = get_fft1(np.conj(arr_fft))
    return np.conj(arr_ifft) / len(arr_fft)
```

#### Inverse fast Fourier transform 2D

```python
import numpy as np


def get_ifft2(img_fft, shift=False):
    h = img_fft.shape[0]
    w = img_fft.shape[1]
    img_fft = img_fft.astype(np.complex)
    
    u_arr = np.arange(h)
    v_arr = np.arange(w)
    
    u_mat = np.repeat(u_arr.reshape((h, 1)), w, 1)
    v_mat = np.repeat(v_arr.reshape((1, w)), h, 0)
     
    if shift:
        exp_val = (u_mat + v_mat) * np.pi * 1j
        img_fft = np.multiply(img_fft, np.exp(exp_val))

    img_ifft = np.conj(get_fft2(np.conj(img_fft))) / (h*w)
    return img_ifft
```

We can also show each *processed* chunk of an image the FFT algorithm process at a certain iteration.

```python
import numpy as np
import matplotlib.pyplot as plt


img_test = np.zeros((128, 128), dtype=float)
img_test[::2, ::2] = 1
img_test[::3, ::3] = 3
img_test[::11, ::11] = 5
plt.figure(figsize=(10, 10), dpi=100)
plt.matshow(img_test, cmap='gray')
plt.show()

fft_vis = get_fft2(img_test, False, True)
```

The 1st image is the original test image generated above, while the last image is the final result of the FFT.
The whole process can be described as follows: first of all, the algorithm picks first two rows of the original image, finds their DFT, then picks the next two rows and finds their DFT, then the algorithm combines these two rows into four rows; further, the process repeats for another two pairs of two-rows, which then combine into four rows, which is followed by combining the two four-rows into one eight-row.

The recursive nature of the algorithm is therefore very clear: it starts from "the depth" (two-rows) to the "upper level" (H-rows, where H is the height of an image).

## Homomorphic filtering
### Theory

Another technique of image processing is homomorphic filtering.
The idea behind it is to think of image as of a combination of illumination $i(x, y)$ and reflectance $r(x, y)$:
$$
\begin{equation}
    f(x, y) = i(x, y)r(x, y).
    \tag{2}
\end{equation}
$$
In order to perform operations in frequency domain, another type of such decomposition is exploited:
$$
\begin{equation}
    z(x, y) = \ln f(x, y) = \ln i(x, y) + \ln r(x, y),
    \tag{3}
\end{equation}
$$
so that $Z(u, v) = \mathfrak{F}[z(x, y)]$ is also a simple sum of $\ln i$ and $\ln r$ in frequency domain.
We denote $\mathfrak{F}[\ln i(x, y)]$ by $F_i(u, v)$.
$F_r(u, v)$ stands for the same denotion for $r(x, y)$.
With that in mind,
$$
\begin{equation}
    Z(u, v) = F_i(x, y) + F_r(x, y).
    \tag{4}
\end{equation}
$$
After a filter $H(u, v)$ is applied, we can obtain a transformed image in spatial domain as follows:
$$
\begin{equation}
    g(x, y) = e^{s(x, y)} = e^{i'(x, y)} e^{r'(x, y)},
    \tag{5}
\end{equation}
$$
where $i'(x, y) = \mathfrak{F}^{-1}[H(u, v)F_i(u, v)]$ and $r'(x, y) = \mathfrak{F}^{-1}[H(u, v)F_r(u, v)]$.

In the current work, we exploit a modified Gaussian high pass filter as a homomorphic one:
$$
\begin{equation}
    H(u, v) = (\gamma_H - \gamma_L)\left[ 1 - e^{-c(D^2(u, v)/D_0^2)} \right] + \gamma_L,
    \tag{6}
\end{equation}
$$
where $c$ and $D_0$ are constants, $D(u, v)$ returns a value of distance of $(u, v)$ from the center of an image, $\gamma_H > 1$ and $\gamma_L < 1$ -- parameters, which control how filter affects **h**igh frequency (which correspond to reflections) and **l**ow frequency (corresponding to external lightning) parts of an image.

### Code

#### Gaussian homomorphic filter

```python
import numpy as np


def apply_gs_hom(img_fft, g_h, g_l, c=1., d0perc=0.5):
    h = img_fft.shape[0]
    w = img_fft.shape[1]
        
    d0 = d0perc * np.min([h, w]) / 2
    
    u_arr = np.arange(h)
    v_arr = np.arange(w)
    
    u_mat = np.repeat(u_arr.reshape((h, 1)), w, 1)
    v_mat = np.repeat(v_arr.reshape((1, w)), h, 0)
    
    dist_mat = np.sqrt(np.power((u_mat - h//2), 2) + np.power((v_mat - w//2), 2))
    
    filter_mat = (g_h - g_l) * (np.full_like(dist_mat, 1.) - np.exp(-c*np.power(dist_mat, 2)/(d0*d0))) + \
                 g_l * np.full_like(dist_mat, 1.)
    
    img_fft = np.multiply(img_fft, filter_mat)
    
    return img_fft
```
